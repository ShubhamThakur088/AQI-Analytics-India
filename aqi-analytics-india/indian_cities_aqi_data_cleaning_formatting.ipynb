```python
import numpy as np

# Creating a 
# aqi_index_city_wise = pd.DataFrame()
#& (aqi_data_frame['date'].dt.year == 2025)

# converting 'date' column to proper datetime. dayfirst is to suppress warning of yyyy-mm-dd format
aqi_data_frame['date'] = pd.to_datetime(aqi_data_frame['date'], dayfirst = True)

# Filter for city & year to get aqi_values of the same.
aqi_index_city_wise_2024 = aqi_data_frame.loc[(aqi_data_frame['area'] == 'Hyderabad') & (aqi_data_frame['date'].dt.year == 2024) , 'aqi_value']
aqi_index_city_wise_2023 = aqi_data_frame.loc[(aqi_data_frame['area'] == 'Hyderabad') & (aqi_data_frame['date'].dt.year == 2023) , 'aqi_value']

# Calculating mean/average of each year
print(aqi_index_city_wise_2023.mean())
print(aqi_index_city_wise_2024.mean())

# Creating X-plot and Y-plot using numpy. With years on x-axis and city-wise, year-wise mean of aqi_values
x_points = np.array(['2023', '2024'])
y_points = np.array([aqi_index_city_wise_2023.mean(), aqi_index_city_wise_2024.mean()])

# Providing labels to X-axis and Y-axis
plt.xlabel('year')
plt.ylabel('Yearly Avg Of AQI')

plt.plot(x_points, y_points)
plt.show()
```
